```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import random
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE

```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

from sklearn.metrics import precision_score,recall_score, f1_score,accuracy_score, confusion_matrix
from sklearn.metrics import roc_auc_score,roc_curve,auc,classification_report
```

Preparing columns for the model

```python
df_cc['time_scaled']=sc.fit_transform(df_cc[['Time']])
df_cc['amount_scaled']=sc.fit_transform(df_cc[['Amount']])
```

```python
X=df_cc.drop(['Time','Amount','Class'],axis=1)
y=df_cc['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=50, stratify=y)
```

##### Basic model with the imbalanced data

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_predict_lr_imb=lr.predict(X_test)
```

```python
print("Accuracy:",accuracy_score(y_test, y_predict_lr_imb))
print("Precision:",precision_score(y_test, y_predict_lr_imb))
print("Recall:",recall_score(y_test, y_predict_lr_imb))
print("f1:",f1_score(y_test, y_predict_lr_imb))
print("ROC AUC Score:",roc_auc_score(y_test, y_predict_lr_imb))
fpr, tpr, thresholds =roc_curve(y_test, y_predict_lr_imb)
roc_auc = auc(fpr, tpr)
print(fpr)
print(tpr)
print(thresholds)
```

```python
cm=confusion_matrix(y_test, y_predict_lr_imb)
sns.heatmap(cm, annot=True,fmt = ".1f",
                    cmap=plt.cm.Reds,
                   xticklabels=['No Fraud', 'Fraud'],
                   yticklabels=['No Fraud', 'Fraud']).set_title('Logistic Regression')
```

```python
print(classification_report(y_test, y_predict_lr_imb))
```

```python
pickle.dump(lr, open('../models/lr_imb.pkl', 'wb'))
```

XGBoost with tuning scale-pos-weight for imbalance to match the classweights.
scale_pos_weight is number of samples in majority class/number of samples in minority class

```python
y_train.value_counts()
```

```python
weights = [1, 25, 50, 100, 200, 300, 400, 500]
param_grid = dict(scale_pos_weight=weights)
grid_search = GridSearchCV(xgb, param_grid, scoring="f1_macro", n_jobs=-1, cv=5)


grid_result = grid_search.fit(X_train, y_train)
```

```python
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
xgb_pos=XGBClassifier(scale_pos_weight=100)
xgb_pos.fit(X_train, y_train)
```
